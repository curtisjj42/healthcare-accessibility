```python
hospital_info = big_df['Hospital Information']
snf_info = big_df['Nursing home information']
```

```python
import re
def normalize_string(s):
    return re.sub(r'[^\w\s]', '', s).lower()
hospital_info['norm_address'] = hospital_info['Address'].apply(normalize_string)
prov_ma_df['norm_name'] = prov_ma_df['Provider Name'].apply(normalize_string)
snf_info['norm_address'] = snf_info['Address'].apply(normalize_string)
```

```python
results_df = prov_ma_df[prov_ma_df['norm_address'].isin(snf_info['norm_address'])]
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

ma_map = gpd.read_file('CENSUS2020_BLK_BG_TRCT/CENSUS2020TRACTS_POLY.shp')
ma_map = ma_map.to_crs('EPSG:4326')
print(ma_map.crs)
ma_map.plot()
```

```python
gdf = gpd.GeoDataFrame(
    results_df, 
    geometry=gpd.points_from_xy(results_df['Longitude'], results_df['Latitude']),
    crs="EPSG:4326")
```

```python
fig, ax = plt.subplots()
ma_map.plot(ax=ax, color="lightgrey", edgecolor="black")

# Overlay the points on top of the shapefile
gdf.plot(ax=ax, color="red", markersize=50)

# Display the plot
plt.show()
```

```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="myGeocoder")
def geocode_address(address):
    try:
        location = geolocator.geocode(address)
        return location.latitude, location.longitude if location else (None, None)
    except Exception as e:
        return (None, None)
snf_info[['latitude', 'longitude']] = snf_info['Address'].apply(lambda x: geocode_address(x)).apply(pd.Series)
```
